```python
def get_property(raw):
    lofl = {}
    for k,v in rawx.items():
        lofl[str(k)] = len(v["vibration-x"])
    t= sorted(lofl.items(),key=lambda x:x[1])
    minl = t[0]
    maxl = t[-1]
    df = pd.DataFrame(lofl,index=["lenofx"])
    df = df.transpose()
    return df,df["lenofx"].value_counts()
```

```python
def min_max_filter(df,prop):
    total = prop.values.sum()
    prop = prop.to_frame()
    prop["ratio"] = prop["lenofx"] / total
    prop["cumlative"] = prop["ratio"].cumsum()
    prop["dev"] = np.abs(prop.index - prop.index[0])
    selected = prop[((prop["cumlative"]<0.9) | (prop["dev"] <=350))&(prop["lenofx"]>=20)]
    result=df[df["lenofx"].isin(selected.index.values)]
    return result
    
```
